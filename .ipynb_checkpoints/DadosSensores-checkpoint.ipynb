Análise dos Dados dos Sensores

```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.model_selection import train_test_split
```

```python
# Caminhos para os arquivos
data_folder = "Dados"

# Listar arquivos de dados brutos de sensores
sensor_files = [f for f in os.listdir(data_folder) if f.endswith('.txt')]

print("\nArquivos de dados brutos de sensores:")
print(sensor_files)
```

```python
# Criar dicionário para armazenar os dados amostrados
sensor_data_sampled = {}

# Iterar sobre cada sensor
for file in sensor_files:
    # Carregar os dados do sensor
    df_sensor = pd.read_csv(f"{data_folder}//{file}", delimiter="\t", header=None)
    
    # Adicionar a coluna de ciclo
    df_sensor["Ciclo"] = df_sensor.index + 1

    # Transformar os dados para formato longo
    df_long = df_sensor.melt(id_vars=["Ciclo"], var_name="Tempo", value_name="Valor")

    # 1️⃣ Amostragem Estratificada: Garante distribuição equilibrada das condições do sistema
    df_stratified, _ = train_test_split(df_long, test_size=0.9, stratify=df_long["Ciclo"], random_state=42)

    # 2️⃣ Amostragem Sistemática: Escolhe ciclos espaçados dentro da amostra estratificada
    k = 2  # Exemplo: escolhe a cada 2 ciclos
    df_sample = df_stratified.iloc[::k, :]

    # Armazenar os dados amostrados
    sensor_data_sampled[file] = df_sample

    # Exibir resultado da amostragem
    print(f"{file}: Amostragem final contém {df_sample['Ciclo'].nunique()} ciclos selecionados.")
```
