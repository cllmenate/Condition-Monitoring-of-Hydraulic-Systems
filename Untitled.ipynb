```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics  as sts
from scipy.stats import f_oneway
from scipy.stats import pearsonr, spearmanr
```

```python
#importar dados
dataset = pd.read_csv("dataset_hydraulicSystems.csv", sep=",")
#visulizar
dataset.head()
```

```python
# Definir as variÃ¡veis-alvo
target_cols = ["Cooler_condition", "Valve_condition", "Pump_leakage", "Accumulator_pressure", "Stable_flag"]

# Criar grÃ¡ficos de distribuiÃ§Ã£o para cada variÃ¡vel-alvo
fig, axes = plt.subplots(1, 5, figsize=(20, 5))

for i, col in enumerate(target_cols):
    sns.countplot(x=dataset[col], ax=axes[i], palette="viridis")
    axes[i].set_title(f"DistribuiÃ§Ã£o de {col}")
    axes[i].set_xlabel("Classe")
    axes[i].set_ylabel("FrequÃªncia")

plt.tight_layout()
plt.show()
```
