```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics  as sts
from scipy.stats import f_oneway
from scipy.stats import pearsonr, spearmanr
from sklearn.preprocessing import MinMaxScaler
```

```python
#importar dados
dataset = pd.read_csv("dataset_hydraulicSystems.csv", sep=",")
#visulizar
dataset.head()
```

```python
# Definir as variÃ¡veis-alvo
target_cols = ["Cooler_condition", "Valve_condition", "Pump_leakage", "Accumulator_pressure", "Stable_flag"]

# Criar grÃ¡ficos de distribuiÃ§Ã£o para cada variÃ¡vel-alvo
fig, axes = plt.subplots(1, 5, figsize=(20, 5))

for i, col in enumerate(target_cols):
    sns.countplot(x=dataset[col], ax=axes[i], palette="viridis", hue=dataset[col], legend=False)
    axes[i].set_title(f"DistribuiÃ§Ã£o de {col}")
    axes[i].set_xlabel("Classe")
    axes[i].set_ylabel("FrequÃªncia")

plt.tight_layout()
plt.show()
```

```python
for col in target_cols:
    print(dataset[col].value_counts())
```

```python
# Selecionar apenas as colunas de sensores e variÃ¡veis-alvo
sensor_cols = [col for col in dataset.columns if col not in ["Ciclo"] + target_cols]
dataset_corr = dataset[sensor_cols + target_cols].corr()

# Criar um heatmap da matriz de correlaÃ§Ã£o
plt.figure(figsize=(12, 8))
sns.heatmap(dataset_corr[target_cols], annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("CorrelaÃ§Ã£o entre Sensores e VariÃ¡veis-Alvo")
plt.show()
```

```python
# Definir o limiar de correlaÃ§Ã£o mÃ­nima (exemplo: Â±0.1)
correlation_threshold = 0.1

# Selecionar sensores
sensor_cols = [col for col in dataset.columns if col not in ["Ciclo"] + target_cols]

# Calcular a correlaÃ§Ã£o absoluta com as variÃ¡veis-alvo
low_corr_sensors = []
for sensor in sensor_cols:
    max_corr = dataset_corr[sensor][target_cols].abs().max()  # Pega a maior correlaÃ§Ã£o com qualquer variÃ¡vel-alvo
    if max_corr < correlation_threshold:
        low_corr_sensors.append(sensor)

# Mostrar sensores pouco relevantes
print("Sensores com baixa correlaÃ§Ã£o e que podem ser removidos:")
print(low_corr_sensors)
```

```python
dataset_reduced = dataset.drop(columns=low_corr_sensors)
dataset_reduced.to_csv("dataset_filtrado.csv", index=False)

print(f"Dataset reduzido: {dataset_reduced.shape[1]} colunas restantes")
```

```python
# Gerar estatÃ­sticas das variÃ¡veis numÃ©ricas
dataset_reduced.describe()
```

```python
# Selecionar algumas variÃ¡veis para plotar
amostra_sensores = ['PS1_media', 'TS1_media', 'VS1_media', 'EPS1_media']

# Criar grÃ¡ficos de distribuiÃ§Ã£o
fig, axes = plt.subplots(1, len(amostra_sensores), figsize=(15, 5))
for i, sensor in enumerate(amostra_sensores):
    sns.histplot(dataset_reduced[sensor], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f"DistribuiÃ§Ã£o de {sensor}")

plt.tight_layout()
plt.show()
```

```python
# Definir um limite de outliers usando IQR
def remove_outliers_iqr(dataset, columns):
    dataset_clean = dataset.copy()
    for col in columns:
        Q1 = dataset[col].quantile(0.25)
        Q3 = dataset[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Remover outliers
        dataset_clean = dataset_clean[(dataset_clean[col] >= lower_bound) & (dataset_clean[col] <= upper_bound)]
    
    return dataset_clean

# Aplicar o filtro de outliers
sensor_cols = [col for col in dataset_reduced.columns if col not in target_cols]  # Apenas sensores
dataset_outliers_removed = remove_outliers_iqr(from sklearn.preprocessing import MinMaxScaler, sensor_cols)

print(f"Dataset original: {dataset.shape[0]} amostras")
print(f"Dataset apÃ³s remoÃ§Ã£o de outliers: {dataset_outliers_removed.shape[0]} amostras")
```

```python
# Aplicar normalizaÃ§Ã£o (0 a 1)
scaler = MinMaxScaler()
dataset_normalized = dataset_outliers_removed.copy()
dataset_normalized[sensor_cols] = scaler.fit_transform(dataset_outliers_removed[sensor_cols])

print("Dados normalizados com sucesso!")
```

```python
# Selecionar algumas variÃ¡veis para plotar
amostra_sensores = ['PS1_media', 'TS1_media', 'VS1_media', 'EPS1_media']

# Criar grÃ¡ficos de distribuiÃ§Ã£o
fig, axes = plt.subplots(1, len(amostra_sensores), figsize=(15, 5))
for i, sensor in enumerate(amostra_sensores):
    sns.histplot(dataset_normalized[sensor], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f"DistribuiÃ§Ã£o de {sensor}")

plt.tight_layout()
plt.show()
```

```python
window_size = 5  # Tamanho da janela deslizante

dataset_moving_avg = dataset_normalized.copy()
for col in sensor_cols:
    dataset_moving_avg[f"{col}_mov_avg"] = dataset_normalized[col].rolling(window=window_size, min_periods=1).mean()

print("MÃ©dias mÃ³veis aplicadas com sucesso!")
```

```python
dataset_derivatives = dataset_moving_avg.copy()
for col in sensor_cols:
    dataset_derivatives[f"{col}_deriv"] = dataset_moving_avg[col].diff().fillna(0)

print("Derivadas calculadas com sucesso!")
```

```python
# Selecionar alguns sensores para anÃ¡lise
amostra_sensores = ['PS1_media', 'TS1_media', 'VS1_media', 'EPS1_media']

# Criar grÃ¡ficos comparativos
fig, axes = plt.subplots(len(amostra_sensores), 2, figsize=(12, 10))

for i, sensor in enumerate(amostra_sensores):
    # GrÃ¡fico da MÃ©dia MÃ³vel
    axes[i, 0].plot(dataset_normalized[sensor], label='Original', alpha=0.5)
    axes[i, 0].plot(dataset_moving_avg[f"{sensor}_mov_avg"], label='MÃ©dia MÃ³vel', color='red')
    axes[i, 0].set_title(f'MÃ©dia MÃ³vel - {sensor}')
    axes[i, 0].legend()

    # GrÃ¡fico da Derivada
    axes[i, 1].plot(dataset_derivatives[f"{sensor}_deriv"], label='Derivada', color='green')
    axes[i, 1].set_title(f'Derivada - {sensor}')
    axes[i, 1].legend()

plt.tight_layout()
plt.show()
```

```python
# Calcular a matriz de correlaÃ§Ã£o
correlation_matrix = dataset_derivatives.corr()

# Focar na correlaÃ§Ã£o com as variÃ¡veis-alvo
target_correlation = correlation_matrix[target_cols]

# Plotar o heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(target_correlation, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("CorrelaÃ§Ã£o das Features com as VariÃ¡veis-Alvo")
plt.show()
```
